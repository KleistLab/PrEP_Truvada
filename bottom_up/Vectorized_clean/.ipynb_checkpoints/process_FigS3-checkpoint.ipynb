```python
from PEPV import *
from matplotlib import pyplot as plt
import torch
import matplotlib 
import numpy as np
```

```python
from math import factorial
a = 0.0996
def f(r):
    res = 0
    for i in range(80):
        res += factorial(84) / factorial(i) / factorial(84 - i) * r**i * (1 - r) **(84 - i) * (1 - (1-a)**i)
    return res
```

```python
t0, t1 = 0, 480
r = Regimen('DTG', 24, (t0, t1), 0, 3, 1)
e = EfficacyPredictor()
e.add_regimen(r)
e.compute_extinction_probability()
e.compute_cumulative_extinction_probability()
pe = e.get_extinction_probability()
cp = e.get_cumulative_extinction_probability()
```

```python
import matplotlib
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=11)
fig, ax = plt.subplots(1, 3, figsize=(12, 3))
plt.subplots_adjust(wspace=0.3)
n = 12000
t = [i*0.02/24 for i in range(int(0), int(10 * 24/0.02))]
ax[0].plot(t, pe[:n, 0, 0, 0, 0]*100, lw=1.5, c='crimson', label='$P_E(\hat{V})$')
ax[0].plot(t, pe[:n, 0, 0, 1, 0]*100, lw=1.5, c='forestgreen',label='$P_E(\hat{T_1})$')
ax[0].plot(t, pe[:n, 0, 0, 2, 0]*100, lw=1.5, c='dodgerblue',label='$P_E(\hat{T_2})$')
ax[0].set_xticks([0, 2, 4, 6, 8, 10])
ax[0].set_xlabel('Time of viral exposure [day]')
ax[0].set_ylabel('Extinction probability [\%]')
ax[0].legend(loc=[0.55, 0.1])

t1 = [i/24 for i in range(0, 240)]
ax[1].plot(t1, torch.diff(cp[:241, 0, 0, 0, 0], dim=0), lw=1.5, c='gold', label= '1 virus')
ax[1].plot(t1, torch.diff(cp[:241, 0, 0, 0, 0]**10, dim=0), lw=1.5, c='indigo', label= '10 viruses')
ax[1].set_xlabel('Days after viral exposure')
ax[1].set_xticks([0, 2, 4, 6, 8, 10])
ax[1].set_ylabel('$f(X)$')
ax[1].legend(loc=[0.5, 0.2])

n = np.array([3e-5, 5e-5, 8e-5, 1e-4, 2e-4, 3e-4, 5e-4, 
     8e-4, 1e-3, 2e-3, 3e-3, 5e-3, 8e-3, 1e-2, 2e-2, 3e-2, 5e-2, 8e-2, 1e-1])
ax[2].plot(n * 84, [f(r) for r in n])
r1 = 3.7136e-3
r2 = 3.6266e-4
ax[2].plot(r1*84, f(r1), 'go', label='Homosexual')
ax[2].plot(r2*84, f(r2), 'yo', label='Heterosexual')
ax[2].set_xscale('log')
ax[2].set_xlabel('$E[V_{t_0}]$')
ax[2].set_ylabel('Average infection probability per exposure')
ax[2].set_xticks([1e-2, 1e-1, 1,  10])
ax[2].get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax[2].legend(loc=[0.05, 0.7])
plt.savefig('figS3.svg', format='svg')
```
